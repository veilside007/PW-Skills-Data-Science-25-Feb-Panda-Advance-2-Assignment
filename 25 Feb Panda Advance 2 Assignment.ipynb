```python
Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?

The `loc` and `iloc` are both accessors in pandas DataFrame used for indexing and selecting subsets of data, but they differ in how they select data:

1. **`loc`:**
   - `loc` is primarily label-based and is used to access a group of rows and columns by labels.
   - It allows you to select data based on the row and column labels.
   - It can accept boolean arrays as well, allowing for more flexible selection criteria.
   - The syntax for using `loc` is `df.loc[row_label, column_label]`.

2. **`iloc`:**
   - `iloc` is primarily integer position-based and is used to access a group of rows and columns by integer position.
   - It allows you to select data based on the row and column indices.
   - It only accepts integer inputs and cannot accept boolean arrays.
   - The syntax for using `iloc` is `df.iloc[row_index, column_index]`.

In summary, `loc` is used to access data based on labels (index and column names), while `iloc` is used to access data based on integer positions. Here's a comparison:

# Example usage of loc
df.loc[0]  # Selects the first row based on label
df.loc[:, 'course_name']  # Selects the 'course_name' column based on label

# Example usage of iloc
df.iloc[0]  # Selects the first row based on integer position
df.iloc[:, 0]  # Selects the first column based on integer position

In general, if you are dealing with labeled data and want to select subsets of data based on labels, you should use `loc`. If you are dealing with positional or numerical index-based data and want to select subsets of data based on integer positions, you should use `iloc`.
```

```python
Q6. What do you understand about the windows function in pandas and list the types of windows
functions?

In pandas, window functions are operations that operate on a set of rows defined by a window. These functions allow you to perform calculations on a subset of data defined by a sliding or rolling window. Window functions are especially useful for calculating statistics and aggregations over rolling or expanding windows of data.

Types of window functions in pandas include:

1. **Rolling Window Functions:**
   - Rolling window functions operate on a fixed-size window of consecutive rows and slide over the DataFrame.
   - They are useful for calculating moving statistics such as rolling mean, rolling sum, rolling standard deviation, etc.
   - Example functions: `rolling.mean()`, `rolling.sum()`, `rolling.std()`, `rolling.min()`, `rolling.max()`.

2. **Expanding Window Functions:**
   - Expanding window functions calculate statistics over all data points from the start of the DataFrame up to the current row.
   - They include cumulative statistics such as cumulative sum, cumulative product, etc.
   - Example functions: `expanding.sum()`, `expanding.mean()`, `expanding.std()`, `expanding.min()`, `expanding.max()`.

3. **Exponentially Weighted Window Functions:**
   - Exponentially weighted window functions assign exponentially decreasing weights to data points as you move away from the current row.
   - They are useful for calculating exponentially weighted moving statistics such as exponentially weighted moving average, exponentially weighted moving standard deviation, etc.
   - Example functions: `ewm.mean()`, `ewm.std()`, `ewm.var()`, `ewm.min()`, `ewm.max()`.

These window functions provide powerful tools for analyzing and understanding time-series data, allowing you to efficiently compute various statistics and aggregations over different types of window sizes and configurations.
```

```python
Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
calculates the difference between them in days, hours, and minutes using Pandas time delta. The
program should prompt the user to enter the dates and display the result.

Here's a Python program that takes two dates as input in the format YYYY-MM-DD, calculates the difference between them in days, hours, and minutes using Pandas timedelta, and then displays the result:

import pandas as pd

def calculate_time_difference(start_date, end_date):
    # Convert input strings to datetime objects
    start_datetime = pd.to_datetime(start_date)
    end_datetime = pd.to_datetime(end_date)
    
    # Calculate time difference using Pandas timedelta
    time_difference = end_datetime - start_datetime
    
    # Extract days, hours, and minutes from the timedelta object
    days = time_difference.days
    hours = time_difference.seconds // 3600
    minutes = (time_difference.seconds % 3600) // 60
    
    return days, hours, minutes

# Prompt the user to enter the dates
start_date = input("Enter the start date (YYYY-MM-DD): ")
end_date = input("Enter the end date (YYYY-MM-DD): ")

# Calculate the time difference
days, hours, minutes = calculate_time_difference(start_date, end_date)

# Display the result
print("Time difference:")
print("Days:", days)
print("Hours:", hours)
print("Minutes:", minutes)

This program prompts the user to enter two dates in the format YYYY-MM-DD, then calculates the difference between them using Pandas timedelta. Finally, it extracts the number of days, hours, and minutes from the timedelta object and displays the result.
```

```python
Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then display the sorted data.

Here's a Python program that reads a CSV file containing categorical data, converts a specified column to a categorical data type, prompts the user to enter the file path, column name, and category order, and then displays the sorted data:

import pandas as pd

def convert_to_categorical(df, column_name, category_order):
    # Convert the specified column to categorical data type
    df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)
    return df

# Prompt the user to enter the file path
file_path = input("Enter the file path of the CSV file: ")

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Prompt the user to enter the column name and category order
column_name = input("Enter the column name to convert to categorical: ")
category_order = input("Enter the category order (comma-separated): ").split(',')

# Convert the specified column to categorical data type
df = convert_to_categorical(df, column_name, category_order)

# Display the sorted data
sorted_data = df.sort_values(by=[column_name])
print("Sorted Data:")
print(sorted_data)

This program prompts the user to enter the file path of the CSV file, the column name to convert to a categorical data type, and the category order as a comma-separated list. It then reads the CSV file into a DataFrame, converts the specified column to a categorical data type using the provided category order, sorts the data based on the specified column, and displays the sorted data.
```

```python
Q10. Write a Python program that reads a CSV file containing sales data for different products and
visualizes the data using a stacked bar chart to show the sales of each product category over time. The
program should prompt the user to enter the file path and display the chart.

To accomplish this task, you can use pandas for reading the CSV file and manipulating the data, and then use matplotlib or seaborn for visualizing the data using a stacked bar chart. Below is the Python program that reads a CSV file containing sales data for different products and visualizes the data using a stacked bar chart:

import pandas as pd
import matplotlib.pyplot as plt

def visualize_sales_data(file_path):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Convert 'Date' column to datetime format
    df['Date'] = pd.to_datetime(df['Date'])

    # Group the data by 'Date' and 'Product Category', and sum the 'Sales' column
    sales_data = df.groupby(['Date', 'Product Category'])['Sales'].sum().unstack(fill_value=0)

    # Plot the stacked bar chart
    sales_data.plot(kind='bar', stacked=True, figsize=(10, 6))
    plt.title('Sales of Each Product Category Over Time')
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.xticks(rotation=45)
    plt.legend(title='Product Category')
    plt.tight_layout()
    plt.show()

# Prompt the user to enter the file path
file_path = input("Enter the file path of the CSV file: ")

# Visualize the sales data using a stacked bar chart
visualize_sales_data(file_path)

This program prompts the user to enter the file path of the CSV file containing sales data. It then reads the CSV file into a DataFrame, converts the 'Date' column to datetime format, groups the data by 'Date' and 'Product Category', and sums the 'Sales' column for each group. Finally, it plots the data as a stacked bar chart using matplotlib and displays the chart.
```

```python
Q11. You are given a CSV file containing student data that includes the student ID and their test score. Write
a Python program that reads the CSV file, calculates the mean, median, and mode of the test scores, and
displays the results in a table.
The program should do the followingM
I Prompt the user to enter the file path of the CSV file containing the student dataR
I Read the CSV file into a Pandas DataFrameR
I Calculate the mean, median, and mode of the test scores using Pandas toolsR
I Display the mean, median, and mode in a table.
Assume the CSV file contains the following columnsM
I Student ID: The ID of the studentR
I Test Score: The score of the student's test.
Example usage of the program:
Enter the file path of the CSV file containing the student data: student_data.csv
+-----------+--------+
| Statistic | Value |
+-----------+--------+
| Mean | 79.6 |
| Median | 82 |
| Mode | 85, 90 |
+-----------+--------+
Assume that the CSV file student_data.csv contains the following data:
Student ID,Test Score
1,85
2,90
3,80
4,75
5,85
6,82
7,78
8,85
9,90
10,85
The program should calculate the mean, median, and mode of the test scores and display the results
in a table.

You can achieve this task using pandas to read the CSV file and calculate the mean, median, and mode of the test scores. Here's the Python program:

import pandas as pd

def calculate_statistics(file_path):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)
    
    # Calculate mean, median, and mode of the test scores
    mean_score = df['Test Score'].mean()
    median_score = df['Test Score'].median()
    mode_score = df['Test Score'].mode().tolist()

    return mean_score, median_score, mode_score

def display_statistics(mean_score, median_score, mode_score):
    # Create a DataFrame to display the statistics
    statistics_df = pd.DataFrame({
        'Statistic': ['Mean', 'Median', 'Mode'],
        'Value': [mean_score, median_score, ', '.join(map(str, mode_score))]
    })

    print(statistics_df.to_string(index=False))

# Prompt the user to enter the file path of the CSV file
file_path = input("Enter the file path of the CSV file containing the student data: ")

# Calculate statistics
mean_score, median_score, mode_score = calculate_statistics(file_path)

# Display statistics
print("+-----------+--------+")
print("| Statistic | Value  |")
print("+-----------+--------+")
display_statistics(mean_score, median_score, mode_score)
print("+-----------+--------+")

This program prompts the user to enter the file path of the CSV file containing the student data. It then calculates the mean, median, and mode of the test scores using pandas tools and displays the results in a table format.
```
